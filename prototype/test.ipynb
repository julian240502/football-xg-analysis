```python
for _, row in competitions.iterrows():
    comp_id = row["competition_id"]
    season_id = row["season_id"]
    if row["competition_name"] == "Ligue 1" and row["country_name"] == "France":
        comp_matches = sb.matches(competition_id=comp_id, season_id=season_id)
        paris_matches = comp_matches[
            (comp_matches["home_team"].str.contains("Paris", case=False, na=False)) |
            (comp_matches["away_team"].str.contains("Paris", case=False, na=False))
        ]
```

```python
events_df = sb.events(match_id=3829500)
events_df.head(5)
```

```python
#separate start and end locations from coordinates
events_df[['x', 'y']] = events_df['location'].apply(pd.Series)
events_df[['pass_end_x', 'pass_end_y']] = events_df['pass_end_location'].apply(pd.Series)
events_df[['carry_end_x', 'carry_end_y']] = events_df['carry_end_location'].apply(pd.Series)
```
